```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import csv
 
user_agent = UserAgent()
data = []
 
product_urls = [
    'https://www.snapdeal.com/products/pressure-cookers?sort=plrty#bcrumbLabelId:2831',
    'https://www.snapdeal.com/products/cookware-sets?sort=plrty#bcrumbLabelId:2830',
    'https://www.snapdeal.com/products/pots-pans?sort=plrty#bcrumbLabelId:2831',
    'https://www.snapdeal.com/products/tawa-kadhai?sort=plrty#bcrumbLabelId:2838',
    'https://www.snapdeal.com/products/kadhai-woks?sort=plrty#bcrumbLabelId:2834',
    'https://www.snapdeal.com/products/steamer-idli-maker?sort=plrty#bcrumbLabelId:46104985',
    'https://www.snapdeal.com/products/barbeque-accessories?sort=plrty#bcrumbLabelId:2832',
    'https://www.snapdeal.com/products/dinnerware?sort=plrty#bcrumbLabelId:205',
    'https://www.snapdeal.com/products/glasses-jugs?sort=plrty#bcrumbLabelId:2826',
    'https://www.snapdeal.com/products/serveware?sort=plrty#bcrumbLabelId:2833',
    'https://www.snapdeal.com/products/cutlery?sort=plrty#bcrumbLabelId:2835',
    'https://www.snapdeal.com/products/dining-serving-bowls?sort=plrty#bcrumbLabelId:46104979',
    'https://www.snapdeal.com/products/dining-serving-plates?sort=plrty#bcrumbLabelId:46104995',
    'https://www.snapdeal.com/products/bakeware-moulds-tins?sort=plrty#bcrumbLabelId:2839',
    'https://www.snapdeal.com/products/baking-tools-accessories?sort=plrty#bcrumbLabelId:46104992',
    'https://www.snapdeal.com/products/home-kitchen-tea-coffee-serveware?sort=plrty#bcrumbLabelId:5',
    'https://www.snapdeal.com/products/cups-saucers-tea-sets?sort=plrty#bcrumbLabelId:2856',
    'https://www.snapdeal.com/products/kettles-carafes?sort=plrty#bcrumbLabelId:2857'
]
 
for main_url in product_urls:
    r = requests.get(main_url, headers={'user-agent': user_agent.chrome})
    soup = BeautifulSoup(r.text, 'html.parser')
    results = soup.find_all('div', attrs={'class': 'product-tuple-description'})
 
    for result in results:
        name = result.find('p', attrs={'class': 'product-title'})
        MRP = result.find('span', attrs={'class': 'lfloat product-desc-price strike'})
        Discount_price = result.find('span', attrs={'class': 'lfloat product-price'})
 
        if name is not None and MRP is not None and Discount_price is not None:
            detail = {}
            detail['Title'] = name.text
            detail['MRP'] = MRP.text
            detail['Discount_price'] = Discount_price.text
            data.append(detail)
print(data)
 
filename = 'cooking_utensils.csv'
 
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f, ['Title', 'MRP', 'Discount_price'])
    w.writeheader()
 
    for detail in data:
        w.writerow(detail)
```
